# This is a simple Keras Deep Nueral Network looking at Kansas City Housing Data to try and predict the price of the houses.

## The dataset is located here: https://www.kaggle.com/shivachandel/kc-house-data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/kc_house_data.csv')
```

```python
df.corr()['price'].sort_values()
```

```python
df = df.drop('id',axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].apply(lambda date:date.month)
df['year'] = df['date'].apply(lambda date:date.year)
df = df.drop('date',axis=1)
```

```python
df.columns
```

```python
df = df.drop('zipcode',axis=1)
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

Scale after the split because we only want to scale the train data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x= X_train, y=y_train, validation_data=(X_test,y_test),
         batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
explained_variance_score(y_test,predictions)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Redefine model and train it now with early stopping
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#for a binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# with dropout
#Redefine model and train it now with early stopping
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(.5))

#for a binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```
